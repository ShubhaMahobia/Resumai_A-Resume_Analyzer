```python
import spacy
from spacy.training.example import Example
import random
```

```python

# Fresh model load karo
nlp = spacy.blank("en")  # Blank English model se shuru karo taaki conflicts na ho
ner = nlp.add_pipe("ner")  # NER pipe explicitly add karo

# Labels add karo
TRAIN_DATA = [
    # Sample 1: Skills Section
    (
        "Skills: C++, Machine Learning, Tableau",
        {
            "entities": [
                (0, 6, "SECTION"),     # "Skills"
                (8, 11, "SKILL"),      # "C++"
                (13, 28, "SKILL"),     # "Machine Learning"
                (30, 37, "SKILL")      # "Tableau"
            ]
        }
    ),
    
    # Sample 2: Experience Section
    (
        "Experience: Web Developer at Amazon, 2018-2020",
        {
            "entities": [
                (0, 10, "SECTION"),     # "Experience"
                (12, 25, "JOB_TITLE"),  # "Web Developer"
                (29, 35, "COMPANY"),    # "Amazon"
                (37, 46, "DATE")        # "2018-2020"
            ]
        }
    ),
    
    # Sample 3: Education Section
    (
        "Education: M.Sc in Data Science, MIT, 2016-2018",
        {
            "entities": [
                (0, 9, "SECTION"),       # "Education"
                (11, 15, "DEGREE"),      # "M.Sc"
                (19, 31, "FIELD"),       # "Data Science"
                (33, 36, "INSTITUTION"), # "MIT"
                (38, 47, "DATE")         # "2016-2018"
            ]
        }
    ),
    
    # Sample 4: Certifications Section
    (
        "Certifications: AWS Certified Solutions Architect, 2021",
        {
            "entities": [
                (0, 14, "SECTION"),     # "Certifications"
                (16, 49, "SKILL"),      # "AWS Certified Solutions Architect"
                (51, 55, "DATE")        # "2021"
            ]
        }
    ),
    
    # Sample 5: Programming Skills
    (
        "Programming: JavaScript, Ruby, PostgreSQL",
        {
            "entities": [
                (0, 11, "SECTION"),    # "Programming"
                (13, 23, "SKILL"),     # "JavaScript"
                (25, 29, "SKILL"),     # "Ruby"
                (31, 41, "SKILL")      # "PostgreSQL"
            ]
        }
    ),
    
    # Sample 6: Experience with Project
    (
        "Experience: AI Engineer at Tesla, 2020-2023, Project: Autopilot",
        {
            "entities": [
                (0, 10, "SECTION"),     # "Experience"
                (12, 23, "JOB_TITLE"),  # "AI Engineer"
                (27, 32, "COMPANY"),    # "Tesla"
                (34, 43, "DATE"),       # "2020-2023"
                (53, 61, "PROJECT")     # "Autopilot"
            ]
        }
    ),
    
    # Sample 7: Soft Skills
    (
        "Soft Skills: Teamwork, Problem Solving, Communication",
        {
            "entities": [
                (0, 11, "SECTION"),     # "Soft Skills"
                (13, 20, "SKILL"),      # "Teamwork"
                (22, 36, "SKILL"),      # "Problem Solving"
                (38, 51, "SKILL")       # "Communication"
            ]
        }
    ),
    
    # Sample 8: Roles and Responsibilities (Naya Section)
    (
        "Roles and Responsibilities: Develop web applications, Manage databases",
        {
            "entities": [
                (0, 26, "SECTION"),         # "Roles and Responsibilities"
                (28, 52, "RESPONSIBILITY"), # "Develop web applications"
                (54, 70, "RESPONSIBILITY")  # "Manage databases"
            ]
        }
    ),
    
    # Sample 9: Roles and Responsibilities with Company
    (
        "Roles and Responsibilities: Analyze data at Infosys, 2019-2021",
        {
            "entities": [
                (0, 26, "SECTION"),         # "Roles and Responsibilities"
                (28, 40, "RESPONSIBILITY"), # "Analyze data"
                (44, 51, "COMPANY"),        # "Infosys"
                (53, 62, "DATE")            # "2019-2021"
            ]
        }
    ),
    
    # Sample 10: Mixed Roles and Skills
    (
        "Roles and Responsibilities: Code in Python, Lead team projects",
        {
            "entities": [
                (0, 26, "SECTION"),         # "Roles and Responsibilities"
                (28, 40, "RESPONSIBILITY"), # "Code in Python"
                (42, 47, "SKILL"),          # "Python"
                (49, 65, "RESPONSIBILITY")  # "Lead team projects"
            ]
        }
    )
]

for _, annotations in TRAIN_DATA:
    for ent in annotations.get("entities"):
        ner.add_label(ent[2])

# Optimizer initialize karo
optimizer = nlp.begin_training()

# Training loop
for i in range(20):
    random.shuffle(TRAIN_DATA)
    losses = {}
    for text, annotations in TRAIN_DATA:
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annotations)
        nlp.update([example], drop=0.5, sgd=optimizer, losses=losses)

# Model save karo
nlp.to_disk("resume_ner")

# Test karo
nlp = spacy.load("resume_ner")
test_text = "Skills: Python, TensorFlow, Experience: Analyst at Google, 2021-2023"
doc = nlp(test_text)
for ent in doc.ents:
    print(ent.text, ent.label_)
```
